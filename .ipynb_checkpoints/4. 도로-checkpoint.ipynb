# 6) 정규화

### 1) 혼잡시간강도

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
traffic_time = pd.read_csv("data/4. 도로/(동 추가처리완료)혼잡시간강도.csv")
traffic_time
```

```python
pd.set_option('display.max_rows', None)
traffic_time
```

```python
traffic_time["혼잡시간강도"] = traffic_time["혼잡시간강도"].astype(float)
```

```python
traffic_time.dtypes
```

```python
# X = traffic_time["혼잡시간강도"]

# X_scaled = min_max_scaler.fit_transform(X)
# X_scaled

# traffic_time = pd.DataFrame(X_scaled, column="혼잡시간강도")
# traffic_time
```

```python
max = traffic_time["혼잡시간강도"].max()
min = traffic_time["혼잡시간강도"].min()

traffic_time["정규화"] = 0.01

for i in range(len(traffic_time)):
                 traffic_time["정규화"][i] = (traffic_time["혼잡시간강도"][i] - min)/(max -min)
```

```python
traffic_time
```

```python
traffic_time = traffic_time.set_index(['시군구명','읍면동명'])
```

```python
traffic_time.to_csv("data/4. 도로/(정규화 최종본)혼잡시간강도.csv")
```
