```python
import requests
header = {"Authorization":"KakaoAK ab7339ccf7dc64734e1acaef6ba81767"}
### kakaoMap 함수
def kakaoMap(name_list):
    address = []
    lat = []
    lng = []
    for name in name_list:
        try:
            url = "https://dapi.kakao.com/v2/local/search/keyword.json?query={}".format(name)
            place = requests.get(url, headers=header).json()["documents"]
    
            address.append(place[0].get("address_name"))
            lat.append(place[0].get("y"))
            lng.append(place[0].get('x'))
    
#             print(name + "---->" + place[0].get("address_name")) 
        except:
            address.append("ERROR")
            lat.append("None")
            lng.append("None")
            print(name+": ERROR")
    return address,lat,lng
```

## 버스정류장 / seoul_bus

```python
seoul_bus.head()
```

```python
### 좌표 string 생성
lat_lng_set = []
i = 0
seoul_bus["Y좌표"][0]
for x in seoul_bus["X좌표"]:
    y = seoul_bus["Y좌표"][i]
    lat_lng_set.append(" ".join((str(y),",",str(x))))
    i =i+1
lat_lng_set
```

```python
## 지오코딩을 이용한 주소 추출
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address
address_list = []
for i in lat_lng_set:
    address = geocoding_reverse(i)
    address_list.append(address)
address_list
```

```python
## 주소중 동만 추출
import re
dong = []

for i in address_list:
    try:
        dong_data = re.search("\S+동,",i[0])
        dong.append(dong_data.group(0))
    except:
        dong.append("None")
len(dong)
```

```python
##동에서 ,빼기
dong_final = []

for i in dong :
    dong_final.append(i.split(",")[0])
dong_final
```

```python
df = pd.DataFrame([dong_final]).T
df.columns = ["행정동"]
seoul_bus["행정동"] = df["행정동"]
seoul_bus = seoul_bus[["정류소명","X좌표","Y좌표","행정동"]]
seoul_bus
```

```python
seoul_bus.to_csv("data/2. 공간정보/전처리데이터/버스정류장(정류소명,좌표,행정동).csv")
```
