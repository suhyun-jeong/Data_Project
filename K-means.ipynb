```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("data/QGIS_전처리/충전소추천_격자_위경도.csv")
df
```

```python
X = df[["lon","lat"]]
X
```

```python
model = KMeans(n_clusters=170).fit(X)
```

```python
labels = []
for i in model.labels_:
    labels.append(i)
labels
```

```python
label = pd.DataFrame([labels]).T
label.columns = ["라벨"]
label.value_counts().head(20)
```

```python
df["라벨"] = label["라벨"]
```

```python
df_save = pd.DataFrame()
for i in (52,2,22,36,45,84,77,4,44,25,23,127,5,144,29,26,1,54,119,7):
    df_save = df_save.append(df[df["라벨"] == i])
df_save = df_save[["gid","라벨"]]
df_save["라벨"].value_counts()
```

```python
df_save.to_csv("data/QGIS_전처리/상위20개.csv", index=False)
```
