```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:\\Windows\\Fonts\\malgun.ttf").get_name()
plt.rc("font", family=font_name)

import matplotlib as mlp
mlp.rcParams["axes.unicode_minus"] = False

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("data/QGIS_전처리/최종데이터_인구수전처리전.csv")
df
```

```python
### NaN 0으로 변환
df = df.fillna(0)
df
```

```python
### 정수로 변환
df["급속"] = df["급속"].astype(int)
df["완속"] = df["완속"].astype(int)
df
```

```python
df = df.drop("시군구명", axis=1)
df
```

```python
## 인구수 정규화
num_max = df["인구수"].max()
num_min = df["인구수"].min()
df["인구수"] = (df["인구수"]-num_min)/(num_max-num_min)
df
```

```python
df.to_csv("data/QGIS_전처리/도시계획결합전_데이터.csv")
```

# 충전소 유무 칼럼 추가

```python
df = pd.read_csv("data/최종데이터.csv")
df
```

```python
charge_check = []
for i in df["충전소개수"]:
    if(i>0):
        charge_check.append(1)
    else:
        charge_check.append(0)
charge_check
```

```python
df["충전소유무"] = charge_check
df
```

```python
df.to_csv("data/최종데이터.csv")
```
